```python
# Q1> Write a function to get num ,char, spe_char from any file 
# make generalize fucntion 

def file_read(fn):
    fp = open(fn,"r")
    data = fp.read()
    return data

def filter_content(fd):
    num =[]
    char = []
    spe_char = []
    for word in fd:
        if word.isalnum():
            if word.isnumeric():
                num.append(word)
            else:
                char.append(word)
        else:
            spe_char.append(word)
    return num,char,spe_char

def main():
    fn ='python_sample.txt'
    fd = file_read(fn)
    no,ch,sp_char = filter_content(fd)

    print("Numeric:","".join(no))
    print("Char:","".join(ch))
    print("Special:","".join(sp_char))


if __name__=='__main__':
    main()


# *Note: join() is a string method.
# It joins all elements of an iterable (like a list or tuple) into one single string,
# using the string before .join() as the separator.
```

```python
# Assigment:
# File -- mobilen nos
# File --> Emails --> valid and invalid 

def file_read(fn):
    fp = open(fn,"r")
    data = fp.readlines()    
    fp.close()
    return data

def filter_content(fd):
    valid_emails = []
    invalid_emails = []
    mobile_numbers = []

    for line in fd[1:]:   # skip header // skip the first line
        parts = line.strip().split(",")
        if len(parts) < 7:
            continue

        emp_id, emp_name, salary, email_id, age, address, mobile_no = parts[:7]

        # Email check
        if "@" in email_id and ".com" in email_id:
            valid_emails.append(email_id)
        else:
            invalid_emails.append(email_id)

        # Mobile check (just check length = 10)
        if len(mobile_no) == 10 and mobile_no.isdigit():
            mobile_numbers.append(mobile_no)

    return valid_emails, invalid_emails, mobile_numbers

         
    

def main():
    fn ='employees_sample_data.csv'
    fd = file_read(fn)
    valid_emails, invalid_emails, mobile_numbers = filter_content(fd)
    
    print("Valid Emails:", valid_emails[:5])
    print("Invalid Emails:", invalid_emails[:5])
    print("Mobile Numbers:", mobile_numbers[:5])

    

if __name__ == "__main__":
    main()
```

```python
# Q2> Write a function to find below info from any file (meta data) 
# file:
# 1.num of lines
# 2.num of chars
# 3.num of words
# 4.num of special char
# 5.num of numbes


def file_info(filename):
    fd = open(filename,encoding="utf-8")
    data= fd.read()

    num_lines = data.count('\n') +1
    num_chars = len(data)
    words =data.split()
    num_words = len(words)
   
    num_special = 0
    num_digits = 0

    for ch in data:
        if not ch.isalnum() and not ch.isspace():
            num_special +=1
        if ch.isdigit():
            num_digits +=1
            
    print("File Name :", filename)
    print("1 Number of lines      :", num_lines)
    print("2 Number of characters :", num_chars)
    print("3 Number of words      :", num_words)
    print("4 Number of special chars:", num_special)
    print("5 Number of numbers    :", num_digits)
    

def main():
    fn = "python_sample.txt" 
    file_info(fn)

if __name__ == "__main__":
    main()
```

```python
# Q1> Assignment :
# mobile -- collect -- list

def file_read(fn):
    fp = open(fn, "r")
    data = fp.readlines()
    fp.close()
    return data


def filter_content(fd):
    valid_emails = []
    invalid_emails = []
    mobile_numbers = []

    with (
        open("files/valid_emails.txt", "a") as f_valid_email,
        open("files/invalid_emails.txt", "a") as f_invalid_email,
        open("files/valid_mobiles.txt", "a") as f_valid_mobile
    ):
        

        for line in fd[1:]:  # skip header // skip the first line
            parts = line.strip().split(",")
           

            emp_id, emp_name, salary, email_id, age, address, mobile_no = parts[:7]

            # Email check
            if "@" in email_id and ".com" in email_id:
                valid_emails.append(email_id)
                f_valid_email.write(email_id + "\n")
            else:
                invalid_emails.append(email_id)
                f_invalid_email.write(email_id + "\n")

            # Mobile check (just check length = 10)
            if len(mobile_no) == 10 and mobile_no.isdigit():
                mobile_numbers.append(mobile_no)
                f_valid_mobile.write(mobile_no + "\n")

    return valid_emails, invalid_emails, mobile_numbers


def main():
    fn = 'files/employees_sample_data.csv'
    fd = file_read(fn)
    valid_emails, invalid_emails, mobile_numbers = filter_content(fd)

    print("Valid Emails:", valid_emails[:5])
    print("Invalid Emails:", invalid_emails[:5])
    print("Mobile Numbers:", mobile_numbers[:5])
    print("File is genreated")


if __name__ == "__main__":
    main()
```

```python
# Q> Write a program to read machine fail code file
#    and filter on 1st Alphabet/char and create file for A–Z code

def file_read(fn):
    fp = open(fn, "r")
    data = fp.readlines()
    fp.close()
    return data


def filter_content(fd):
    for s in map(str.strip, fd):  # strip spaces/newlines
        if not s:  # skip empty lines
            continue

        first = s[0].upper()  # get first character (make uppercase)

        # If code starts with A–Z → write to that file
        if first.isalpha():
            file_name = "files/machine_code/machine_failcode_{}.txt".format(first)
            
        else:
            # Otherwise write to 'other' file
            file_name = "files/machine_code/machine_failcode_other.txt"

        # Write directly (append mode)
        with open(file_name, "a") as fo:
            fo.write(s + "\n")

    print(" All files generated successfully!")


def main():
    fn = "files/machine_fail_codes.txt"
    fd = file_read(fn)
    filter_content(fd)


if __name__ == "__main__":
    main()
```

```python
# normal open vs open with

# 1>normal open:
#   fp =open(filename)
#   fp.close() -> manually 
# -- we have to manually close 

# 2>with open:
#     with open(filename,mode) as fp:

# -- we don't need to close file manually



#-- normal open

fp = open("files/sample.txt")
data =  fp.read()
print("Before close:",fp.closed)
fp.close()
print("After close:",fp.closed)
```

```python
# -- with open
with open("files/sample.txt","r") as fp:
    data = fp.read()
    print("Before close:",fp.closed)

print("After close:",fp.closed)
```

```python
#4> read then write-- r+

# - sequence is important of read and write
# - it will read then read content of file and the write data in file at the end of file pointer/cursor

with open("files/sample.txt","r+") as fp:
    data = fp.read()
    print(data)
    fp.write("this is content by r+ mode")
```

```python
#5> write then read --w+

# - it will write first and then read
# - so write will first delete the content of files if files has already content then it will delete it
# - so in read mode the pointer of file come to end and no output shown / blank 

with open("sample.txt","w+") as fp:
    fp.write("this is demo for w+ mode")
    data = fp.read()
    print(data)


# -- so you will get no output because file pointer is at last/end

# -- to overcome this issue we can use seek(0)
# -- it will put file pointer at the start of file
# -- tell() put file pointer at the end of file


with open("sample.txt","w+") as fp:
    fp.write("this is demo for w+ mode")
    print(fp.tell())
    fp.seek(0)
    print(fp.tell())
    data = fp.read()
    print(data)
```

```python
#6> append then read- a+

# -- it will first append and then read
# --it will append content where last file pointer is there
# --the pointer go to last and the write and read

with open("sample.txt","a+") as fp:
    fp.write("\nthis is the demo for a+")
    fp.seek(0)
    data = fp.read()
    print(data)
```
