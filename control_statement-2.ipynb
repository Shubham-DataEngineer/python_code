```python
#using if esle ladder


capital_Alphabet = []
small_Alphabet = []
numbers = []
symbols = [] 


for char in p:
    if char.isupper():
        capital_Alphabet.append(char)
    elif char.islower():
        small_Alphabet.append(char)
    elif char.isdigit():
        numbers.append(char)
    else:
        symbols.append(char)

print("capital_Alphabet: ",capital_Alphabet)
print("small_Alphabet: ",small_Alphabet)
print("numbers: ",numbers)
print("symbols: ",symbols)

        
    
    
        
    
```

```python
# Loop control statement
#1> break
# it will terminate the loop
# it is mainly used for validation purpose


eid = [4,54,78,484,22,33,77]
for item in eid:
    if str(item).isnumeric():
        print("process: {}",format(eid))
    else:
        break
print("outside the loop!!!!")





```

```python
#2> continue
# it will skip and directly go to next itreation 




eid = [4,54,78,484,'abc',22,'xyz',33,77]
for item in eid:
    if str(item).isnumeric():
        print("process: {}",format(eid))
    else:
        continue
print("outside the loop!!!!")


```
